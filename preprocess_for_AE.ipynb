```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from datetime import datetime

# sklearn
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn import metrics

# keras
import keras
from keras import models, optimizers
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten, LSTM
from keras import Model ,models, layers, optimizers, regularizers
from keras.callbacks import ModelCheckpoint
import keras.backend.tensorflow_backend as K
import tensorflow as tf
```

```python
data = pd.read_csv("dataset/total2.csv", index_col=2)

# 라벨 처리
data.loc[data['Label'] != 'Benign', "Label"] = 1
data.loc[data['Label'] == 'Benign', "Label"] = 0
data = data.apply(pd.to_numeric, errors = 'coerce')

#결측치 제거
data = data.replace([np.inf, -np.inf], np.nan)
data.fillna(0.0,inplace=True)

# 정상만 학습 (학습_정상 70%, 탐지_정상+악성 30%)
train, test = train_test_split(data, test_size=0.3)
train, valid = train_test_split(train, test_size=0.3)

# 학습용 라벨 분리
train_label = train['Label'].astype('float64').values
valid_label = valid['Label'].astype('float64').values
test_label = test['Label'].astype('float64').values

valid = valid.drop('Label',axis=1).values
test = test.drop('Label', axis=1).values

# 정상만 분리
train_normal = train[train_label == 0].drop('Label',axis=1).values
valid_normal = valid[valid_label == 0].drop('Label',axis=1).values


# min max 정규화
scaler = MinMaxScaler(feature_range=(0,1))
train_normal_scaled = scaler.fit_transform(train_normal)
valid_normal_scaled = scaler.fit_transform(valid_normal)
valid_scaled = scaler.fit_transform(valid)
test_scaled = scaler.fit_transform(test)

# train, valid, test 데이터 저장
np.save('dataset/train_normal_scaled_AE.npy', train_normal_scaled)
np.save('dataset/valid_normal_scaled_AE.npy', valid_normal_scaled)
np.save('dataset/valid_scaled_AE.npy', valid_scaled)
np.save('dataset/test_scaled_AE.npy', test_scaled)


# 라벨 데이터 저장
np.save('dataset/valid_label_AE.npy', valid_label)
np.save('dataset/test_label_AE.npy', test_label)
```
